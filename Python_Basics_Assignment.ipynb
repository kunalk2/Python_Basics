```python
# Q6. Write a function to demonstrate the behavior of mutable and immutable arguments.
def demo(immutable, mutable):
    print("Inside function before changes:")
    print("immutable:", immutable)
    print("mutable:", mutable)

    # Modify the immutable argument
    immutable += 10
    # Modify the mutable argument i.e. adds to the list in this case
    mutable.append(10)

    print("\nInside function after changes:")
    print("immutable:", immutable)
    print("mutable:", mutable)

x = 5          # immutable Object i.e. integer
y = [5]        # mutable Object i.e list
print("Before function call:")
print("x:", x)
print("y:", y)

demo(x, y)

print("\nAfter function call:")
print("x:", x)
print("y:", y)
# here x will will remain unchanged as its an immutable object
```

```python
# Q14. Write a program to reverse a string using a while loop.
String = "PYTHON"
Reversed_String = ""
i = len(String) - 1

while i >= 0:
    Reversed_String += String[i]
    i -= 1

print("Reversed string:", Reversed_String)
```

```python
# Q15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
Number = int(input("Enter a number: "))
factorial = 1
i = 1

# Calculate factorial using 'while' loop
while i <= Number:
    factorial *= i
    i += 1

print("Factorial of", Number, "is:", factorial)
```
